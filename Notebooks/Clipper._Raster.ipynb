```python
#import required libraries
import rasterio
from rasterio import plot
from rasterio.mask import mask
import matplotlib.pyplot as plt
import fiona
from shapely.geometry import shape
import os
%matplotlib inline
```

```python
import pyproj
from shapely.ops import transform
```

```python
path = "/home/ziad/Documents/Github/ShallowLearn/Data/Honduras_Square.shp"
```

```python
with fiona.open(path, "r") as shapefile:
    shapes =[shape(feature["geometry"]) for feature in shapefile]
```

```python
shapes
```

```python
def cut_polys(fname, input_path, output_path, transform_type):
    # read imagery file
    with rasterio.open(f"{input_path}") as src:
        poly_crs = pyproj.CRS.from_epsg(4326)
        transformer = pyproj.Transformer.from_crs(poly_crs, src.crs, always_xy=True)
        transform_func = transformer.transform
        print(shapes)
        new_polygon = transform(transform_func, shapes[0])
        print(new_polygon)
        out_image, out_transform = rasterio.mask.mask(src,[new_polygon], crop = True)
        out_meta = src.meta
#    Save clipped imagery
    out_meta.update({"driver": "GTiff",
                     "height": out_image.shape[1],
                     "width": out_image.shape[2],
                     "transform": out_transform})

    with rasterio.open(f"{output_path}/{transform_type}_{fname}", "w", **out_meta) as dest:
        dest.write(out_image)
```

```python
for fname in os.scandir(f"/media/ziad/Expansion/Honduras_Extracted/"):
    if (fname.name.endswith("tiff")):
        print(fname.name)
        print(fname.path)
        cut_polys(fname.name, fname.path, "/media/ziad/Expansion/Honduras_Clipped", "no_tf")


```
